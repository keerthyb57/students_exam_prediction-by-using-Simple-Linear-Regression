### insert file.csv

```python
# to Import the pandas libraray
import pandas as pd
```

```python
#to load the dataset file using pandas and to store the file in variable"data"
data=pd.read_csv("student_exam_scores.csv")
#to print the dataset
data
```

```python
#load the hours_studied column and store it in independent
Independent=data[["hours_studied"]]
#print the variable
Independent
```

```python
#load the exam_score column and store it in dependent
Dependent=data[["exam_score"]]
#to print the variable
Dependent
```

### import train_test_split

```python
# from the class model_selection in sci-kit learn we'll import the function train_test_split 
from sklearn.model_selection import train_test_split
```

```python
# to Split the Independent variable using the variable named X_train, X_test
# to split the Dependent varianble using the variable named y_train, y_test
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.20,random_state=0)
```

```python
#to print the variable
x_train
```

```python
#to print the variable
x_test
```

```python
#to print the variable
y_train
```

```python
#to print the variable
y_test
```

### import algorithm

```python
#to import the algorithm named LinearRegression in Sci-kit learn(the class is linear.model)
from sklearn.linear_model import LinearRegression
```

```python
#to store the linear regression algo into an variable named regression
Regression=LinearRegression()
Regression
```

```python
# to check the fit- means for training 
Regression.fit(x_train,y_train)
```

```python
#to find the diff between two points and store the data into weight
weight=Regression.coef_
weight
```

```python
#to find the starting point and store the data into bias
bias=Regression.intercept_
bias
```

```python
# to predict Create the variable named "y_pred" and saved the X_test data in the varibale
y_pred=Regression.predict(x_test)
y_pred
```

### import r2_score for accuracy

```python
#Evaluationm parameters for accuracy-import r2_score from the metrics in sklearn
from sklearn.metrics import r2_score
```

```python
#To find the accuracy score in the variable "r_score"
r_score=r2_score(y_test,y_pred)
r_score
```

### pickle for model saving

```python
#to save the model
import pickle
```

```python
#save the model name and create a file need to end in.sav
file_name="final_examscore_pred.sav"
```

```python
# Save the model into the model name,dump=open,wb=write binary
pickle.dump(Regression,open(file_name,'wb'))
```

```python
#Load the model ,rb-read binary
loaded_model=pickle.load(open("final_examscore_pred.sav",'rb'))
loaded_model
```
